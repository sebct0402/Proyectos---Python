Para verificar los datos atípicos, empecemos con el enfoque estadístico.

Empecemos construyendo diagramas de caja con las variables numéricas.

```python
import math

# Variables numéricas (sin 'player_id')
numericas = df.select_dtypes(include=['number']).columns

# Número de variables
n = len(numericas)

# Definir filas y columnas para la cuadrícula
cols = 3  # Número de columnas (puedes ajustarlo)
rows = math.ceil(n / cols)

# Crear figura y ejes
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 4))
axes = axes.flatten()  # Convertir a lista para iterar fácilmente

# Crear boxplots
for i, col in enumerate(numericas):
    df.boxplot(column=col, ax=axes[i])
    axes[i].set_title(f'Boxplot de {col}')
    axes[i].set_ylabel(col)

# Eliminar ejes vacíos
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

También obtengamos la cantidad exacta de atípicos por cada variable numérica.

```python
# Calculate IQR for each numerical column
Q1 = df[numericas].quantile(0.25)
Q3 = df[numericas].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Count outliers for each numerical column
outliers_count = {}
for col in numericas:
    col_outliers = df[(df[col] < lower_bound[col]) | (df[col] > upper_bound[col])]
    outliers_count[col] = len(col_outliers)

# Display the count of outliers
print("Count of outliers per numerical variable (IQR method):")
for col, count in outliers_count.items():
    print(f"{col}: {count}")
```

##Desarrollo y evaluación de los modelos computacionales


Primero importemos nuestras herramientas de interés

```python
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.preprocessing import OneHotEncoder, StandardScaler, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline as SkPipeline
from imblearn.pipeline import Pipeline  # <- usa imblearn para meter SMOTE en el pipeline
from imblearn.over_sampling import SMOTE

from sklearn.metrics import (
    accuracy_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, ConfusionMatrixDisplay, RocCurveDisplay
)

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

Ahora, para iniciar con el desarrollo de nuestros modelos de clasificación binaria, deberemos verificar el balanceo de las clases.

```python
df['evento_muerte'].value_counts()
```

```python
df['evento_muerte'].value_counts(normalize=True)*100
```

Observamos un desbalance importante, por lo que deberemos aplicar técnicas para el balanceo de clases

Reconociendo que la baja cantidad de datos que tenemos disponibles, debemos tener en cuenta:
- Debemos evitar la pérdida de registros, por lo que SMOTE-Tomek no resulta muy favorable.
- El uso de ADASYN puede generar ejemplos irreales al tener pocos registros.

De esta forma nos decidiremos por SMOTE simple.

Además, para el escalado de datos, probaremos tanto StandardScaler como RobustScaler para identificar cuál nos ofrece un mejor rendimiento.

Empecemos separando nuestro conjunto de datos entre predictores y Target

```python
X = df.drop(columns=["evento_muerte"])
y = df["evento_muerte"]
```

Ahora trabajemos en los pipelines para cada tipo de escalado y tratamiento a las variables.

```python
# Detecta columnas por tipo
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()
```

```python
#Transformaciones de var. numericas para los diferentes scaler
num_transformer_std = SkPipeline(steps=[("scaler", StandardScaler())])
num_transformer_rob = SkPipeline(steps=[("scaler", RobustScaler())])

#Transformación para variables categóricas (Cabe aclarar que es poco probable que esta sección sea usada, pues anteriormente ya se realizó la transformación necesaria)
cat_transformer = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
```

```python
#Ensamblemos nuestro preprocessor para diferentes categorías por cada Scaler
preprocessor_std = ColumnTransformer(
    transformers=[
        ("num", num_transformer_std, num_cols),
        ("cat", cat_transformer, cat_cols),
    ],
    remainder="drop"
)

preprocessor_rob = ColumnTransformer(
    transformers=[
        ("num", num_transformer_rob, num_cols),
        ("cat", cat_transformer, cat_cols),
    ],
    remainder="drop"
)
```

Ensamblemos los pipelines de preproceso en uno sólo.

```python
preprocesadores = {
    "StandardScaler": preprocessor_std,
    "RobustScaler": preprocessor_rob
}
```

Ahora construyamos los modelos que estarán en nuestro pipeline final.

```python
modelos = {
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(
        n_estimators=300, random_state=42, n_jobs=-1
    ),
    "SVM": SVC(kernel="rbf", probability=True, random_state=42)
}
```

Con todo listo, definamos el parámetro de cross validation y algunas estructuras de datos que nos permitan almacenar los resultados.

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

resultados = []           # para tabla de métricas promedio
detalles_confusion = {}   # para matrices de confusión (predicciones OOF)
detalles_prob = {}
```

Con todo esto listo podemos iniciar con el proceso de entremiento de los modelos.

```python
for nombre_prep, prep in preprocesadores.items():
    for nombre_modelo, modelo in modelos.items():

        pipe = Pipeline(steps=[
            ("preprocessor", prep),
            ("smote", SMOTE(random_state=42)),
            ("model", modelo)
        ])
         # Predicciones OOF (out-of-fold) para obtener métricas consistentes
        # Probabilidades para AUC (si el modelo lo soporta)
        try:
            y_proba_oof = cross_val_predict(
                pipe, X, y, cv=cv, method="predict_proba", n_jobs=-1
            )[:, 1]
        except Exception:
            # si no tiene predict_proba, intentamos decision_function y lo normalizamos
            y_dec_oof = cross_val_predict(
                pipe, X, y, cv=cv, method="decision_function", n_jobs=-1
            )
            # Min-max para llevar a [0,1]
            y_min, y_max = y_dec_oof.min(), y_dec_oof.max()
            y_proba_oof = (y_dec_oof - y_min) / (y_max - y_min + 1e-12)

        y_pred_oof = (y_proba_oof >= 0.5).astype(int)

        acc  = accuracy_score(y, y_pred_oof)
        rec  = recall_score(y, y_pred_oof, zero_division=0)
        f1   = f1_score(y, y_pred_oof, zero_division=0)
        auc  = roc_auc_score(y, y_proba_oof)

        resultados.append({
            "Preprocesamiento": nombre_prep,
            "Modelo": nombre_modelo,
            "Accuracy": acc,
            "Recall": rec,
            "F1": f1,
            "AUC": auc
        })

        detalles_confusion[(nombre_prep, nombre_modelo)] = confusion_matrix(y, y_pred_oof)
        detalles_prob[(nombre_prep, nombre_modelo)] = (y, y_proba_oof)
```

Obtengamos nuestra tabla de resultados:

```python
res_df = pd.DataFrame(resultados).sort_values(
    by=["AUC", "F1", "Recall", "Accuracy"], ascending=False
).reset_index(drop=True)

print("Resultados promedio (OOF) ordenados por AUC, luego F1, Recall y Accuracy:")
print(res_df)
```

De esta forma, llegamos a la lista de mejores resultados para todos nuestros escenario.

Podemos tener principalmente 2 conclusiones de estos resultados:
- El método Scaler que mejor resultados aportó fue RobustScaler (para los 4 indicadores).
- El modelo que mejor resultados aportó fue Random Fortes (para los 4 indicadores).

Imprimamos sólamente la mejor combinación

```python
mejor = res_df.iloc[0]
best_prep = mejor["Preprocesamiento"]
best_model = mejor["Modelo"]
print("\n=== Mejor combinación ===")
print(mejor)
```

Obtengamos ahora la correspondiente matriz de confusión y la curva ROC

```python
cm = detalles_confusion[(best_prep, best_model)]
y_true, y_score = detalles_prob[(best_prep, best_model)]

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Matriz de confusión
disp = ConfusionMatrixDisplay(cm, display_labels=[0,1])
disp.plot(ax=ax[0], colorbar=False)
ax[0].set_title(f"Confusión - {best_model} ({best_prep})")

# Curva ROC
RocCurveDisplay.from_predictions(y_true, y_score, ax=ax[1])
ax[1].set_title(f"ROC - AUC={roc_auc_score(y_true, y_score):.3f} - {best_model} ({best_prep})")

plt.tight_layout()
plt.show()
```

Finalmente llegamos a estos resultados:
- El modelo logra clasificar bien el 84% de las observaciones. (Accuracy)
- Para la clase positiva, es decir, los pacientes que fallecieron, el modelo logar identificar correctamente un 78%. (recall)
- Encontramos un balance aceptable entre correcta clasificación de positivos y negativos. (F1 = 0,76)
- Tomando en cuenta pares de pacientes de ambas clasificaciones (fallecido y no fallecido), el 91% de las veces el modelo le dará mayor valor a los pacientes fallecidos. (AUC)

###Conclusiones

Concluimos que el modelo que mejor se comporta (entre los 3 empleados) prediciendo una insuficiencia cardiaca con el conjunto de datos es el de Ramdom Forest entrenado con el escalado de RobustScaler. Además, para el despliegue de un modelo como este en un contexto médico deben ser verificados cuidadosamente los proceso y los resultados de este modelado. Es muy importante prestarle atención a los falsos positivos y verdaderos negativos y las implicaciones éticas que podría desencadenar cada uno de estos escenarios; pues recordemos que están en juego vidas humanas y la tranquilidad de los pacientes.
